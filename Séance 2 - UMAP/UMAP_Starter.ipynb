# Utilisation de UMAP pour la classification de nombre

Notebook pour démarrer le TP, à améliorer. On commence par les librairies principales :

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns; sns.set(style="white")
```

Sklearn offre un dataset *digits* qui correspond à MNIST, mais pas avec la même résolution. Pour accéder à celui en 28x28 pixel, il faut aller dans l'open data :

```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml("mnist_784")
X = mnist.data.to_numpy()
y = mnist.target
```

Visualisons les nombres pour vérifier :

```python
to_display = 8
index_plot = 1

plt.figure(figsize=(16, 8))

for _ in range(to_display):
    index_image = np.random.randint(0, X.shape[0])
    
    plt.subplot(2, 4, index_plot)
    plt.imshow(X[index_image].reshape(28, 28), plt.cm.gray_r, interpolation="nearest")
    plt.title("Image d'un %s" % y[index_image])
    sns.despine()
    
    index_plot += 1

plt.show()
```
